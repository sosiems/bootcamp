```python
import numpy as np
import pandas as pd
import bokeh_catplot
import bokeh.plotting
import bokeh.io
import iqplot

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('data/frog_tongue_adhesion.csv', comment='#')

# Have a look so we remember
df.head()
```

```python
df_mean = df.groupby('ID')['impact force (mN)'].mean().reset_index()

# Take a look
df_mean
```

```python
p = bokeh.plotting.figure(
    frame_height=200,
    frame_width=400,
    x_axis_label='impact force (mN)',
    y_range=df_mean['ID'].unique()[::-1],
    tools='pan,wheel_zoom,save,reset'
)
```

```python
p.hbar(
    source=df_mean,
    y='ID',
    right='impact force (mN)',
    height=0.6
)

# Turn off gridlines on categorical axis
p.ygrid.grid_line_color = None

# Start axes at origin on quantitative axis
p.x_range.start = 0

bokeh.io.show(p)
```

```python
p = bokeh.plotting.figure(
    frame_height=250,
    frame_width=250,
    y_axis_label='impact force (mN)',
    x_range=df_mean['ID'].unique(),
)

p.vbar(
    source=df_mean,
    x='ID',
    top='impact force (mN)',
    width=0.6
)

p.xgrid.grid_line_color = None
p.y_range.start = 0

bokeh.io.show(p)
```

```python
p = bokeh_catplot.box(
    data=df,
    cats='ID',
    val='impact force (mN)'
)

bokeh.io.show(p)
```

```python
p = iqplot.strip(
    data=df,
    cats='ID',
    val='impact force (mN)'
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.histogram(
    data=df,
    cats='ID',
    val='impact force (mN)'
)

bokeh.io.show(p)
```

```python
# Generate normally distributed data
np.random.seed(353926)
df_norm = pd.DataFrame(data={'x': np.random.normal(size=500)})

# Plot the histogram
p = bokeh_catplot.histogram(
    data=df_norm,
    cats=None,
    val='x'
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.ecdf(
    data=df_norm,
    cats=None,
    val='x',
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.ecdf(
    data=df,
    cats='ID',
    val='impact force (mN)'
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.ecdf(
    data=df,
    cats='ID',
    val='impact force (mN)',
    #style='staircase',
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.ecdf(
    data=df,
    cats='ID',
    val='impact force (mN)',
    p=p
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.box(
    data=df,
    cats='ID',
    val='impact force (mN)',
    horizontal=True,
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.box(
    data=df,
    cats='ID',
    val='impact force (mN)',
    whisker_caps=True,
    outlier_marker='diamond',
    box_kwargs=dict(fill_color='#7C0000'),
    whisker_kwargs=dict(line_color='#7C0000', line_width=2),
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.strip(
    data=df,
    cats='ID',
    val='impact force (mN)',
    marker='dash',
    marker_kwargs=dict(alpha=0.5)
)

bokeh.io.show(p)
```

```python
p = bokeh_catplot.strip(
    data=df,
    cats='ID',
    val='impact force (mN)',
    jitter=True,
    tooltips=[
        ('trial', '@{trial number}'),
        ('adh force', '@{adhesive force (mN)}')
    ],
)

bokeh.io.show(p)
```
